```python
import yfinance as yf
import pandas as pd

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
tickers_list = ['EURUSD=X', 'GC=F', 'BTC-USD', 'AAPL', 'XOM', 'V', 'CL=F']

start_date = '2015-01-01'
end_date = '2022-12-31'

data = pd.DataFrame(columns=tickers_list)

for ticker in tickers_list:
    data[ticker] = yf.download(ticker, start_date, end_date)['Close']

data.head()
```

```python
# data.columns = ['EUR', 'GOLD', 'Bitcoin', 'Apple', 'Exxon', 'VISA', 'Oil']
data.info()
```

```python
data.to_csv('../data/mvp_data_close_15-22.csv')
```

```python
plt.figure(figsize=(18, 12))
plt.subplots_adjust(top=1)

for i, company in enumerate(data.columns, 1):
    plt.subplot(3, 3, i)
    data[company].plot()
    plt.xlabel(None)
    plt.title(company)

plt.show()
```
